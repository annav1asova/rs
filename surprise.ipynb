```python
from surprise import Reader, Dataset, SVD, evaluate
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ratings_data = pd.read_csv("/Users/annavlasova/Downloads/purchases.csv")
ratings_data.head()
```

```python
movies = pd.read_csv("/Users/annavlasova/Downloads/content_groups.csv", sep=';')
```

```python
ratings_data['rating'] = 1
```

```python
# Load Reader library
reader = Reader()

# Load ratings dataset with Dataset library
data = Dataset.load_from_df(ratings_data[['user_id', 'uuid_film', 'rating']], reader)

# Split the dataset for 5-fold evaluation
data.split(n_folds=5)
```

```python
# Use the SVD algorithm.
svd = SVD()

# Compute the RMSE of the SVD algorithm.
evaluate(svd, data, measures=['RMSE'])
```

```python
trainset = data.build_full_trainset()
svd.train(trainset)
```

```python
from collections import Counter
inventory = Counter()
```

```python
set_users = ratings_data.user_id.unique()
set_movies = ratings_data.uuid_film.unique()
```

```python
print(len(set_users))
```

```python
users_movies_ns = {"user_id":[], "movies":[]}
users_movies_s = {"user_id":[], "movie":[]}
# users_movies_2 = []
for user_id in set_users:
    r = list(ratings_data.loc[ratings_data['user_id'] == user_id]["uuid_film"])
    if (len(r) == 1):
        users_movies_s["user_id"].append(user_id)
        users_movies_s["movie"].append(r[0])
    elif len(r) == 2 or len(r) == 3:
        pass
#         users_movies_2.append(user_id)
    else:
        users_movies_ns["user_id"].append(user_id)
        users_movies_ns["movies"].append(r)
    loot = {len(r):1}
    inventory.update(loot)
    
    if (len(r) > 50):
        print(len(r), user_id)


print(len(set_users))
print(len(users_movies_ns["user_id"]))
print(len(users_movies_s["user_id"]))

```
